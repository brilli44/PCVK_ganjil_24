3. Implement transformation of logarithmic brightness on Google Colaboratory using the
formula written in the Theoretical Background Reviews, generating output shown in
the following figure:

```python
import cv2 as cv
import numpy as np
import math  # Pastikan modul math diimpor
from google.colab.patches import cv2_imshow

print('Mengubah tingkat kecerahan citra dengan Transformasi Log')
print('--------------------------------------------------------')

# Input nilai kecerahan
try:
    brightness = float(input('Masukkan nilai kecerahan: '))
except ValueError:
    print('Error, not a number')
    brightness = 0  # Set default brightness to 0 if input is invalid

# Baca gambar dari Google Drive
original3 = cv.imread('/content/drive/MyDrive/PCVK/female.tiff')

# Periksa apakah gambar berhasil dibaca
if original3 is None:
    print('Error: Gambar tidak ditemukan!')
else:
    # Buat array kosong untuk gambar hasil
    log_brightness_image = np.zeros(original3.shape, original3.dtype)

    # Akses tiap piksel untuk melakukan transformasi logaritmik
    for y in range(original3.shape[0]):
        for x in range(original3.shape[1]):
            for c in range(original3.shape[2]):
                # Transformasi logaritmik
                log_brightness_image[y, x, c] = np.clip(brightness * math.log(1 + original3[y, x, c]), 0, 255)

    # Gabungkan gambar asli dan gambar hasil untuk ditampilkan berdampingan
    final_frame = cv.hconcat([original3, log_brightness_image])

    # Tampilkan gambar
    cv2_imshow(final_frame)
```

4. implement grayscale transformation using the averaging, lightness, and luminance
method on Google Colaboratory using the formula written in the Theoretical
Background Reviews, generating output as follows :

- **Average**

```python
print('Transformasi Grayscale dengan Averaging')
print('----------------------------------------')

# Baca gambar dari Google Drive
image_path = '/content/drive/MyDrive/PCVK/female.tiff'  # Ganti path dengan path gambar Anda
original = cv.imread(image_path)

# Periksa apakah gambar berhasil dibaca
if original is None:
    print('Error: Gambar tidak ditemukan!')
else:
    # Metode Averaging
    gray_avg = cv.cvtColor(original, cv.COLOR_BGR2GRAY)

    # Ubah grayscale kembali ke format BGR untuk penampilan
    gray_avg_image = cv.cvtColor(gray_avg, cv.COLOR_GRAY2BGR)

    # Gabungkan gambar asli dan grayscale Averaging untuk ditampilkan berdampingan
    final_frame = cv.hconcat([original, gray_avg_image])

    # Tampilkan gambar
    cv2_imshow(final_frame)
```

- **Lightness**

```python
print('Transformasi Grayscale dengan Lightness')
print('----------------------------------------')

# Baca gambar dari Google Drive
image_path = '/content/drive/MyDrive/PCVK/female.tiff'  # Ganti path dengan path gambar Anda
original = cv.imread(image_path)

# Periksa apakah gambar berhasil dibaca
if original is None:
    print('Error: Gambar tidak ditemukan!')
else:
    # Metode Lightness
    min_channel = np.min(original, axis=2)
    max_channel = np.max(original, axis=2)
    gray_lightness = ((max_channel + min_channel) / 2).astype(np.uint8)

    # Ubah grayscale kembali ke format BGR untuk penampilan
    gray_lightness_image = cv.cvtColor(gray_lightness, cv.COLOR_GRAY2BGR)

    # Gabungkan gambar asli dan grayscale Lightness untuk ditampilkan berdampingan
    final_frame = cv.hconcat([original, gray_lightness_image])

    # Tampilkan gambar
    cv2_imshow(final_frame)
```




- **Luminance**

```python
print('Transformasi Grayscale dengan Luminance')
print('----------------------------------------')

# Baca gambar dari Google Drive
image_path = '/content/drive/MyDrive/PCVK/female.tiff'  # Ganti path dengan path gambar Anda
original = cv.imread(image_path)

# Periksa apakah gambar berhasil dibaca
if original is None:
    print('Error: Gambar tidak ditemukan!')
else:
    # Metode Luminance
    # Formula: Y = 0.299*R + 0.587*G + 0.114*B
    gray_luminance = np.dot(original[...,:3], [0.299, 0.587, 0.114]).astype(np.uint8)

    # Ubah grayscale kembali ke format BGR untuk penampilan
    gray_luminance_image = cv.cvtColor(gray_luminance, cv.COLOR_GRAY2BGR)

    # Gabungkan gambar asli dan grayscale Luminance untuk ditampilkan berdampingan
    final_frame = cv.hconcat([original, gray_luminance_image])

    # Tampilkan gambar
    cv2_imshow(final_frame)
```

5.  Show certain colors in the image , and change other colors to grayscale. For example ,
show the blue color in the image input and change the parts other which is not colored
blue into grayscale as in the example below :

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

# Baca gambar dari Google Drive
image_path = '/content/drive/MyDrive/PCVK/female.tiff'  # Ganti path dengan gambar Anda
image = cv.imread(image_path)

# Periksa apakah gambar berhasil dibaca
if image is None:
    print('Error: Gambar tidak ditemukan!')
else:
    # Konversi gambar ke HSV
    hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)

    # Definisikan rentang warna biru dalam HSV
    lower_blue = np.array([100, 150, 0])  # Batas bawah warna biru
    upper_blue = np.array([140, 255, 255])  # Batas atas warna biru

    # Buat mask untuk bagian yang berwarna biru
    mask = cv.inRange(hsv, lower_blue, upper_blue)

    # Pisahkan bagian yang berwarna biru
    blue_part = cv.bitwise_and(image, image, mask=mask)

    # Konversi gambar asli ke grayscale
    gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

    # Ubah grayscale menjadi gambar berwarna untuk gabungan (3 channel)
    gray_3_channel = cv.cvtColor(gray_image, cv.COLOR_GRAY2BGR)

    # Gabungkan bagian biru dengan bagian lainnya yang grayscale
    mask_inv = cv.bitwise_not(mask)
    gray_part = cv.bitwise_and(gray_3_channel, gray_3_channel, mask=mask_inv)
    final_image = cv.add(blue_part, gray_part)

    # Gabungkan gambar asli dan gambar hasil transformasi secara horizontal
    combined_image = cv.hconcat([image, final_image])

    # Tampilkan gambar sebelum dan sesudah
    cv2_imshow(combined_image)
```
