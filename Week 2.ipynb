###**1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?**
- Jawab :
  - **Tanpa `matplotlib` (menggunakan OpenCV)**:OpenCV menampilkan gambar dengan `cv.imshow()` di jendela terpisah, tapi ini tidak berfungsi di Google Colab. Selain itu, OpenCV menggunakan urutan warna BGR, jadi warna mungkin terlihat salah (merah jadi biru, biru jadi merah).
  - **Dengan `matplotlib`**:
  `matplotlib` menampilkan gambar langsung di notebook (seperti Google Colab), jadi lebih cocok digunakan di sana. `matplotlib` juga menggunakan urutan warna RGB yang lebih standar, sehingga warna gambar akan terlihat benar.
  - **Kesimpulan:**
`matplotlib` lebih baik untuk menampilkan gambar di Google Colab, sementara OpenCV cocok untuk aplikasi lokal.

###**2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16 dan int32?**
- **Perbedaan dan pengaruh tipe data `int16` dan `int32` pada pembuatan gambar hitam:**
  - Jwaba : **`int16` (16-bit)**: Menyimpan nilai piksel antara -32,768 hingga 32,767. Rentang ini lebih besar dari yang diperlukan untuk gambar biasa (0-255), jadi menggunakan `int16` untuk gambar standar akan membuang memori dan mungkin memerlukan penyesuaian nilai piksel.
  - **`int32` (32-bit)**: Menyimpan nilai piksel antara -2 miliar hingga 2 miliar, yang juga jauh lebih besar dari yang diperlukan. Seperti `int16`, penggunaan `int32` untuk gambar biasa hanya menghabiskan lebih banyak memori.
  - **Kesimpulan:**
  Untuk gambar biasa dengan nilai piksel 0-255, lebih baik menggunakan tipe data `uint8`. Menggunakan `int16` atau `int32` tidak perlu dan hanya menghabiskan memori lebih banyak.
